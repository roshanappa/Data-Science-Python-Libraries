```python
#we can give the label to the x and y axis and we can give the title to our plot
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
from numpy.random import randint
y=randint(1,50,20)
x= np.linspace(1,10,20)
y =np.sort(y)
plt.plot(x,y,color ='g')
plt.ylabel('y axis')
plt.xlabel('x axis')
plt.title('Simple Plot')
plt.show()
```

```python
#so how use the multiple plots on the same canvas
plt.plot()
#in that pic we having single canvas and by default it having the single axis
plt.subplot(1,2,1)#here we having the subplot command in that we are referring the 1 row and 2 columns and refering the 2nd fig
plt.plot(x,y,'r')
plt.subplot(1,2,2)#here we having the subplot command in that we are referring the 1 row and 2 columns and refering the 1st fig
plt.plot(x,y,'b')
plt.show()
```

### Diffrence between the matlab and matplotlib

```python
#matplotlib as work as matlab as well as it works as objet oriented language or method.its called to figure object
#the above way done with the matlab way.
#now we see the object orienetd way- ploting figure and we will get the control of statics of that fig.
#so we will create figure by object oriited way
fig= plt.figure()
axes = fig.add_axes([0.1,0.1,1,1])
#here x = 0.1,y=0.1 and it will go upto the 1 and 1
#left bottom corner treated as 0.1 corner and top right hand corner treated as 1.
#if we apply the method on it like plot
axes.plot(x,y,'r')
axes.set_xlabel('x axis')
axes.set_ylabel('y axis')
axes.set_title('Set Title Of Random')
plt.show()
```

```python
#if we change anything he x cordinate will see the diffrence in the fig 
#but if we create the one more fig that time it will show the diffrence.
```

```python
#if we want to make the plot inside the plot that time 
#create an objet of figure  
fig = plt.figure()
ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.1,0.6,0.4,0.3])
ax1.plot(x,y,'r')
ax1.set_xlabel('x axis')
ax1.set_ylabel('y axis')
ax2.plot(x,y1,'g')
ax2.set_ylabel('y axis')
ax2.set_xlabel('x axis')
ax1.set_title('Big random plot')
ax2.set_title('Small random plot')
plt.show()
```

### Log-axis-plot

```python
fig,ax = plt.subplots(1,2,figsize = (10,4))
ax[0].plot(x,y,x,y1)
ax[1].plot(x,np.exp(x))
plt.show()
fig.tight_layout()
```

### Set-Scale

```python
#if we check that(y axis) is based on the linear scale so how we can change the linear scale to log sacle of this graph
fig,ax = plt.subplots(1,2,figsize = (10,4))
ax[0].plot(x,y,x,y1)
ax[1].plot(x,np.exp(x))
ax[1].set_yscale('log')
plt.show()
fig.tight_layout()
```
