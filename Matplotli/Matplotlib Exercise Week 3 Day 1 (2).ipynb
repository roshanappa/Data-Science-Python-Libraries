Create a Barplot using following values ?
x1=[3,5,7,9]
y1=[12,15,16,8]
x2=[4,6,8,10]
y2=[15,17,19,10]

```python
x1=[3,5,7,9] 
y1=[12,15,16,8]
x2=[4,6,8,10] 
y2=[15,17,19,10]


plt.barh(x1,y1,label='Data1')
plt.barh(x2,y2,label='Data2')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('horizontal bar graph')
plt.legend()
plt.show()
```

Create a line graph using following list?

```python
x1=[2,3,5,6]
y1=[5,8,10,3]
x2=[5,8,10,3]
y2=[2,3,5,6]
```

```python
plt.plot(x1,y1,color='r',marker ='o',lw=3,label='First Line')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.legend()
plt.grid()
plt.title('line graph')
plt.show()
```

```python
x1=[2,3,5,6]
y1=[5,8,10,3]
x2=[5,8,10,3]
y2=[2,3,5,6]
```

```python
plt.plot(y1,x1)
```

```python
x1=[2,3,5,6]
y1=[5,8,10,3]
x2=[5,8,10,3]
y2=[2,3,5,6]

plt.plot(x1,y1,linestyle='--',color='r',marker='o',lw=2,label='First Line')
plt.plot(x2,y2,marker='D',lw=2,color='b',label='Second Line')
plt.title('Line Graph')
plt.xlabel('First Line')
plt.ylabel('Second Line')
plt.grid()
plt.legend()
plt.show()
```
