#### How the Cufflinks and Plotly works together ?

#### Area Plot


If we want to make the area plot and stack the area plot one over the onther that time we have to use for that we required possitive values.

If we want to make the area plot that time we required either possitive value or negetive value.


```python
#if we kindly check the below graph we having the -ve values those are not really usefull becaue while ploting the with negetive possitive data.that time plot under area would not work.
#Area under plot with the stack that time we required either possitive or negetive.
df = pd.DataFrame(np.random.randn(100,3),columns = ['A','B','C'])
df['A'] = df['A'].cumsum() + 17
df['B'] = df['B'].cumsum() + 17
df['C'] = df['C'].cumsum() + 17

df.iplot()
```

Then make the above plot in the area plot.

We had got the area under the line plot but below that plot we did not get coloured view.

```python
df.iplot(kind='area')
```

For getting the colour to represent the how much area is under the line plot for that we need to use the fill is True.

Apart from that the plot can be toggol by clicking the button right side to get the plot representations of the each category.

```python
df.iplot(kind='area',fill =True)
```

#### Surface Plot With The Help of Datagen Function.

```python
cf.datagen.sinwave(10,0.25).iplot(kind ='surface')
```

#### 3-D Scatter Plot

```python
cf.datagen.scatter3d(2,150,mode ='stocks').iplot(kind='scatter3d',x='x',y='y',z='z')
```

By using this plot we can understand how two variables are changing that means what is spread of two variables.

How they are changing one verses another.

In the cricket team whenver we compare the performaces of the two teams,that shows how they are gaing the performaces one verses another.



```python
df[['A','B']].iplot(kind ='spread')
```
